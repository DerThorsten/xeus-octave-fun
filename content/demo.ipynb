```octave
pkg load image

I = imread("noether.png");
if ndims(I) == 3
  I = rgb2gray(I);
endif
I = im2double(I);

% Gaussian smoothing
I_smooth = imgaussfilt(I, 2);

% edges
edges = edge(I_smooth, "canny");

% gradient
[Gx, Gy] = gradient(I_smooth);
gradmag = sqrt(Gx.^2 + Gy.^2);

% watershed
L = watershed(gradmag);

imshow(label2rgb(L));
title("Watershed segmentation");
```
